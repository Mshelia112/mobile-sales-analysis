```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
import pandas as pd

df = pd.read_csv('synthetic_mobile_sales_2025.csv')
print(f"DataFrame 'df' loaded with {len(df)} rows and {len(df.columns)} columns.")
df.head(50)
```

```python
# Basic info
df.shape
df.info()
df.describe()

# Missing values
df.isnull().sum()

# Duplicates
df.duplicated().sum()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)

# Convert date (if exists)
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

# Price distribution
if 'Price' in df.columns:
    sns.histplot(df['Price'], kde=True)
    plt.title('Price Distribution')
    plt.show()

## Price Distribution Insights
#Most phones are concentrated in the mid-range price category
#Few high-end phones exist, indicating limited premium sales


# Brand count
if 'Brand' in df.columns:
    sns.countplot(x='Brand', data=df)
    plt.xticks(rotation=45)
    plt.title('Brand Distribution')
    plt.show()

## Brand Performance Insights
#Brand X dominates in number of units sold
#Some brands have low sales, suggesting weak market presence


# Correlation heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()

## Relationship Insights
#There is a negative relationship between price and units sold
#Higher-priced phones tend to sell fewer units
```
